```python
# Get the relevant dataframe for analysis
df = df_orig[['price', 'neighbourhood_group_cleansed', 'host_response_time','square_feet','review_scores_rating','review_scores_value','guests_included','beds','bathrooms','host_is_superhost','host_identity_verified','room_type']]
df.head()
```

```python
# Convert price to float 
df.price = df.price.str.strip('$').str.replace(',','')
df.price = df.price.astype(float)
df.host_response_time = df.host_response_time.astype(str)
#df.head()
df.describe
```

```python
# Apply the filters
df = df[df['price'] < 300]
df = df[df['neighbourhood_group_cleansed'].isin(['Centro','Salamanca','Chamberi','Arganzuela','Tetuán','Retiro','Moncloa - Aravaca','Chamartin','Ciudad Lineal','Carabanchel','Latina'])]
df = df[df['room_type'] == 'Entire home/apt']
df = df.drop('room_type', 1) # No need for the analysis
```
